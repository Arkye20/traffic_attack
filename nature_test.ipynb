```python
# from skimage.measure import compare_ssim
from skimage.metrics import structural_similarity as compare_ssim
import cv2
import matplotlib.pyplot as plt
ori = 'yolov3/data/stop-sign.jpg'
adv = 'output_adv/sign/stop-sign.jpg'
img1 = cv2.imread(ori)[:, :, ::-1]
img2 = cv2.imread(adv)[:, :, ::-1]
img1 = cv2.resize(img1,(416,416))
img2 = cv2.resize(img2,(416,416))
ssim = compare_ssim(img1, img2, win_size=11, data_range=255, multichannel=True)
print(ssim)
plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.show()
```
